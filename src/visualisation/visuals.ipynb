```python
import os
import cv2
import matplotlib.pyplot as plt 
import numpy as np

%matplotlib inline

glioma_images, meningioma_images, normal_images, pituitary_tumour_images = [], [], [], []

def get_images(dirname:str, destination:list):
        for image in os.listdir(f'../../data/raw_data/{dirname}'):
                img = cv2.imread(f'../../data/raw_data/{dirname}/{str(image)}')
                destination.append(img)
        destination = np.array(destination)
                

                
get_images("glioma_tumour", glioma_images)
get_images("meningioma_tumour", meningioma_images)
get_images("normal", normal_images)
get_images("pituitary_tumour", pituitary_tumour_images)

```

```python
def show_image(image):
    plt.imshow(image)
    
    
data = iter(glioma_images)
images_to_plot = []

for i in range(20):
    image = glioma_images[i]
    images_to_plot.append(image)
    
figure = plt.figure(figsize=(25, 4))

for i in np.arange(20):
    box = figure.add_subplot(2, 10, i+1)
    show_image(images_to_plot[i])
```
